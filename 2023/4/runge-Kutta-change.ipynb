```python
import numpy as np
import matplotlib.pyplot as plt
import common_module as cm
```

```python
#今回扱う関数の定義
def function(t,y):
      return -1*y + 3

#初期値の設定
t0 = 0
y0 = 1

#刻み幅の設定
dt = 0.1

#計算回数の設定
n = 200000

#計算用の配列を用意
t_sq = [t0]
y_sq = [y0]
```

```python
# ルンゲクッタ法
def RungeKutta(t, y,f):
      k1 = f(t, y)
      k2 = f(t + dt / 2, y + dt / 2 * k1)
      k3 = f(t + dt / 2, y + dt / 2 * k2)
      k4 = f(t + dt, y + dt * k3)
      y_next = y + dt / 6 * (k1 + 2 * k2 + 2 * k3 + k4)
      return y_next
```

```python
t = t0
y = y0

for i in range(n):
      y = RungeKutta(t, y,function)
      t = t + dt
      t_sq = np.append(t_sq, t)
      y_sq = np.append(y_sq, y)
```

```python
#解析解の計算
def analytical_f(t):
    return (y0 - 3) * np.exp(-1*t) + 3
```

```python
#グラフの描画
plt.scatter(t_sq,y_sq,label="Runge-Kutta")
plt.plot(t_sq,analytical_f(t_sq),"r",label="Analytical solution")
plt.xlabel('t')
plt.ylabel('y')
plt.title('Runge-Kutta')
plt.legend()
plt.show()
```
